```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
height = 500
width = 800
img = cv2.imread("ujin1.jpeg")
img = cv2.resize(img, dsize=(width,height), interpolation=cv2.INTER_LINEAR)
img2 = cv2.imread("ujin2.jpeg")
img2 = cv2.resize(img2, dsize=(width,height), interpolation=cv2.INTER_LINEAR)
```

```python
cv2.imshow("img",img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
orb = cv2.ORB_create()
kp1, des1 = orb.detectAndCompute(img,None)
kp2, des2 = orb.detectAndCompute(img2,None)
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck = True)
matches = bf.match(des1,des2)
matches = sorted(matches,key=lambda x:x.distance)

bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2,k=2)
good_points = []
ratio = 0.7
ret = 0
min_dist = 999
for m, n in matches:
    if m.distance < ratio*n.distance:
        if m.distance < min_dist:
            min_dist = m.distance
        good_points.append(m)
        print(len(good_points))
        if m.distance <= 10:
            ret +=1
if ret > 0:
    print("min_dist = " + str(min_dist) + " : same_image")
else:
    print("min_dist = " + str(min_dist) + " : different image")
result = cv2.drawMatches(img, kp1,img2,kp2,good_points,None,flags=2)
plt.imshow(result)
plt.show()
cv2.imwrite('result2.jpeg',result)
```
