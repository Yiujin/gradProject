```python
width1 = cap1.get(cv2.CAP_PROP_FRAME_WIDTH)

height1 = cap1.get(cv2.CAP_PROP_FRAME_HEIGHT)

```

```python
while (cap1.isOpened):

    ret, frame = cap1.read()

 

    if ret:

        cv2.imshow("video", frame)

 

        k = cv2.waitKey(33)

        if k == 27:

            break

    else:

        break

cap1.release()

cv2.destroyAllWindows()
```

```python
while (cap1.isOpened):

    ret, frame = cap1.read()
    
    cv2.imwrite("vid_test%d.jpg" % count,frame)
    
    print('saved frame%d.jpg' % count)
    count +=10
cap1.release()
```

```python
import glob
 
img_array = []
for filename in glob.glob('vid_test*.jpg'):
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)
 
 
out = cv2.VideoWriter('vid_real.avi',cv2.VideoWriter_fourcc(*'DIVX'), 15, size)
 
for i in range(len(img_array)):
    out.write(img_array[i])
out.release()
```
